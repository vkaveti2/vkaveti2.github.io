### **Question 1**: What 20 countries have earned the most Summer Olympic medals? How about the Winter Olympics?

#### Summer

```python
## SUMMER
# Sort data
summer = df[["NOC", "Season", "Medal"]]
summer = summer.loc[summer["Season"] == "Summer"]
summer = summer.dropna(axis='index', subset=["NOC", "Season", "Medal"])
summer
```

```python
most_summer = summer.groupby("NOC").size()
most_summer = most_summer.sort_values(ascending=False)
most_summer = most_summer.head(20)
most_summer
```

```python
# Make the plot
plt.figure(figsize=(12, 8))
most_summer.plot(kind="bar", color = "orange")
plt.title("Top 20 Countries in the Summer Olympics")
plt.xlabel("Country Code")
plt.ylabel("Medal Count")
plt.xticks(rotation=45)
plt.show()
```

This plot shows that the United States has historically earned the most medals in the Summer Olympics, The USSR, Great Britain, Germany, and France are the next countries behind. This graph shows the 20 highest medal getters. Note that the USSR and Russia are counted as separate countries due to their different codes. If you were to combine the medal counts for the USSR and Russia, they would still have the second most Summer medals.


#### Winter

```python
## WINTER
# Sort data
winter = df[["NOC", "Season", "Medal"]]
winter = winter.loc[winter["Season"] == "Winter"]
winter = winter.dropna(axis='index', subset=["NOC", "Season", "Medal"])
winter
```

```python
most_winter = winter.groupby("NOC").size()
most_winter = most_winter.sort_values(ascending=False)
most_winter = most_winter.head(20)
most_winter
```

```python
# Make the plot
plt.figure(figsize=(12, 8))
most_winter.plot(kind="bar", color = "skyblue")
plt.title("Top 20 Countries in the Winter Olympics")
plt.xlabel("Country Code")
plt.ylabel("Medal Count")
plt.xticks(rotation=45)
plt.show()
```

As for the Winter Olympics, the United States has also won the most medals. Canada is very close in second place, with Norway, the USSR, Sweden, and Finland close behind. In the case that the USSR and Russia were counted as the same country, they would actually have the most Winter medals.

### **Question 2:** Have different countries been more successful in different time periods?

```python
# Finds the total number of medals per country (gold, silver, and bronze)
total_medals_per_country = df.groupby('NOC')['Medal'].count().sort_values(ascending=False)
# Selects the top 5 performing countries
top_5_countries = total_medals_per_country.head(5).index
# Finds the data for the top 5 performing countries
top_5_country_data = df[df['NOC'].isin(top_5_countries)]
# Finds the medal count for the top 5 performing countries
top_5_country_year_medals = top_5_country_data.groupby(['NOC', 'Year'])['Medal'].count().reset_index()
plt.figure(figsize=(12, 8))
# Iterates over unique countries in the NOC column
for country in top_5_country_year_medals['NOC'].unique():
# Filters data for the current country
  country_data = top_5_country_year_medals[top_5_country_year_medals['NOC'] == country]
# Creates the plot
  plt.plot(country_data['Year'], country_data['Medal'], label=country)
plt.title('Distribution of Medals Won by Top 5 Performing Countries Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Medals')
plt.legend(loc='upper left', bbox_to_anchor =(1, 1))
plt.tight_layout()
plt.show()
```

The visualization showcases the top 5 countries in terms of medals won throughout the years of the data set (1896 - 2016). URS, the country code for the former Soviet Union, had a large peak in performance around 1980, declined in the next Olympics following the 1980 Olympics, and then spiked back up for the one following that. One can see that all four other countries involved in the visualization have competed in the Olympics every time they have occurred since 1896. URS competed from about 1950 - 1990 and then changed their country code to RUS. From 1990 - 2020, the USA has consistently had the most medals, though France, Great Britain, and the United States all had peaks of high medaling in the early years of the Olympics.


### **Question 3:** Who has the most medals won for women in the summer Olympics?


```python
# Includes only female athletes in the Summer Olympics who have won medals
df_women_summer = df[(df['Sex'] == 'F') & (df['Season'] == 'Summer') & (df['Medal'].notna())]
medal_counts = df_women_summer.groupby(['Name', 'Sport'])['Medal'].count().reset_index(name='Medal Count')
# Sort the medal counts in descending order to find the top medal winners and select the top 11
top_medal_winners = medal_counts.sort_values(by='Medal Count', ascending=False).head(11)
print(top_medal_winners)
```

```python
# Data Visualization code
plt.figure(figsize=(12, 8))
plt.barh(top_medal_winners['Name'], top_medal_winners['Medal Count'], color='purple')
plt.xlabel('Number of Medals')
plt.ylabel('Athlete')
plt.title('Top Female Medal Winners at the Summer Olympics')
plt.gca().invert_yaxis()
plt.show()
```

The visualization shows the 11 highest performing female athletes for the summer Olympics, where many of the women have more than 7 medals apiece.


### **Question 4:** How many appearances did some athletes make?


```python
# Group by name and make the count column Appearances
athlete_count = df.groupby('Name').size().reset_index(name='Appearances')
# Make it descending
athlete_count = athlete_count.sort_values(by='Appearances', ascending=False)
athlete_count.head(10)
```

```python
# Plot the data with the top 10 appearances
plt.figure(figsize=(12, 8))
plt.barh(athlete_count['Name'].head(10), athlete_count['Appearances'].head(10))
plt.xlabel('Number of Appearances')
plt.ylabel('Athlete')
plt.title('Top 10 Athletes by Number of Olympic Appearances')
plt.gca().invert_yaxis()
plt.show()
```

This bar chart measures the top 10 Olympic athlete appearances. Based on the graph, we can see Robert Tait McKenzie has the most appearances resulting in almost 60. Robert Tait McKenzie was an Olympic Sculptor. From 1912 to 1948, the Olympics included fine art competitions as an official event. This graph takes into account the amount of events an athlete competed in and the years they have appeared in.


### **Question 5:** How many countries competed in each Olympics?

```python
# Filter to just year and NOC
year_noc = df[['Year','NOC']]
# Group by year
grouped_year = year_noc.groupby('Year')
# Number of unique Olympic codes (teams) grouped by year
plot = grouped_year.nunique()

plt.figure(figsize=(12, 8))
# Plotting code
plot.plot(kind='bar',color = "skyblue")
plt.title("Number of Teams at Each Olympics")
plt.xlabel("Year")
plt.ylabel("Number of Teams")
```

The bar chart shows the number of teams that compete by the year in each olympics. The data shows a left skew which implies that the olympics grew in popularity slowly over time, and the number of nations competing finally reached 100 in 1968. In the 1990s and after, there appears to be an oscillation between 200+ teams competing and below 100. This visualization shows the drop off between the teams that choose to participate in the summer olympics vs. winter olympics.

### **Question 6:** What are the oldest athletes to win a gold medal? What sports did they play?

```python
# Finds the athletes who have won a gold medal
gold_medalists = df[df['Medal'] == 'Gold']
# Finds the 100 oldest gold medal winners and sorts them from oldest to youngest
oldest_gold_medalists = gold_medalists.sort_values(by = 'Age', ascending = False).head(100)
# Obtains the sports of the oldest athletes and their age
oldest_athletes_sports = oldest_gold_medalists[['Age', 'Sport']]
print(oldest_athletes_sports)
```

```python
# Creates a scatter plot
plt.figure(figsize=(12, 8))
plt.scatter(oldest_athletes_sports['Age'], oldest_athletes_sports['Sport'], color = 'navy', alpha = 0.4)
plt.title('Age and Sport of Oldest Gold Medalists')
plt.xlabel('Age')
plt.ylabel('Sport')
plt.yticks(rotation = 45, ha = 'right')
plt.tight_layout()
plt.show()
```

The darker circles on the visualization represent that there may be multiple individuals who were a specific age in a specific sport who won a gold medal. The oldest individuals to win gold medals were 64 years old and were in Roque and Shooting. There were also a couple individuals who were 63 when they won their gold medals, in Art Competitions and Archery. The data shows that there is not an age limit to winning an Olympic medal, especially since several individuals were winning medals well into their forties and fifties in sports that may require physical endurance.


###**Question 7:** Are there more mens Olympic medals than womens until 1984 (the year when the Olympics changed from every 4 years to every 2 years)? How does the date/time period contribute to this?

```python
# Remove rows where the 'Medal' column contains missing values
df = df.dropna(subset=['Medal'])
# Only include records from the years before 1994
df = df[df['Year'] < 1994]
# Group the data by 'Year', 'Sex', and 'Medal'
# Each medal type is a column
# Fill any missing values with 0 (because no medals of that type were awarded in that group)
medal_counts = df.groupby(['Year', 'Sex', 'Medal']).size().unstack(fill_value=0)
print(medal_counts.head(100))
```

```python
# Visualization code
fig, ax = plt.subplots(figsize=(14, 7))
# This ensures correct formatting since there is a lot of data
medal_counts.plot(kind='bar', stacked=True, ax=ax)
plt.title('Olympic Medals by Sex and Year')
plt.xlabel('Year, Sex')
plt.ylabel('Number of Medals')
plt.legend(title='Medal Type')
plt.show()
```

As we can see from the visualization, there are significantly more medals won by men than women. The time period contributes to this because women were not allowed to compete in the olympics before 1900. Based on the visualization, we can see a general upward trend of all medals, which we can assume is because of the increase of Olympic sports and people.


### **Question 8:** What age has gotten the most gold medals?

```python
# Pull data where the medal is equal to gold
age_gold_medals = df[df['Medal']=='Gold']
# Now, group the previous data by age for the count
gold_counts = age_gold_medals.groupby('Age').size()
gold_counts.head(100)
```

```python
# Plot the data as a bar graph with Age on the x-axis and number of gold medals on the y-axis
plt.figure(figsize=(12, 8))
gold_counts.plot(kind='bar',color ='gold')
plt.xlabel('Age')
plt.ylabel('Number of Gold Medals')
plt.title("Number of Gold Medals Won By Age")
plt.xticks(rotation = 80)
plt.show()
```

Based on this bar chart, we can see a very consistent data peak. The peak seems to be between 22 to 24. The ages that have won the most medals are 23 and 24. It is said that an athlete's physical prime is somewhere between 20 to 30 years old. Based on this data alone, it could be relevant to this statement. The amount of gold medals won diminishes as the age increases.

### **Question 9:** What are the number of Olympics that each athlete competed in?

```python
# Converts years to year format
df['Year'] = pd.to_datetime(df['Year'], format='%Y').dt.year

# Groups athletes by year
athlete_years = df.groupby('Name')['Year'].nunique()

# Plot the data
plt.figure(figsize=(12, 8))
athlete_years.plot(kind="hist")
plt.title("Number of Olympics Competed In")
plt.xlabel("Appearances in Separate Olympic Years")
plt.ylabel("Number of Athletes")
```

The histogram above shows the number of olympians that competed in x amount of olympics. The graph clearly shows that the majority of olympic athletes only get a single chance to compete in the games with just short of 16,000 olympians participating in one olympics.  As we look towards more frequent appearances, there is a significantly large dropoff with around 2,000 olympians with two appearances in separate years and around 400 with 3 appearances in separate years. This graph visualizes the rarity of olympians appearing in more than one olympics across the years.


### **Question 10:** Do medalists often get another medal? Which athletes have the most medals total?

```python
# Finds how many times an athletes name appears in the data frame
only_medalists = df[["Name", "Medal"]]
only_medalists = only_medalists.dropna(axis='index', subset=["Name", "Medal"])
athlete_counts = only_medalists["Name"].value_counts()
# Selects athletes who have won more than one medal
multiple_medalists = athlete_counts[athlete_counts > 1]
# Calculates the percentage of athletes who have won more than one medal
percentage_multiple_medalists = (len(multiple_medalists) / len(athlete_counts)) * 100
rounded_percentage = round(percentage_multiple_medalists, 2)
print(rounded_percentage)
# Finds the names of the athletes and their information
multiple_medalists_names = multiple_medalists.index.tolist()
print(multiple_medalists_names[:10])
```

```python
multiple_medalists_info = df[df['Name'].isin(multiple_medalists_names)]
multiple_medalists_info.head(5)
```

```python
# Finds the count for medals for individuals and sorts them from largest to smallest
medal_counts = multiple_medalists_info.groupby('Name')['Medal'].count()
medal_counts_sorted = medal_counts.sort_values(ascending=False)
# Makes the plot
plt.figure(figsize=(12, 8))
medal_counts_sorted.head(20)[::-1].plot(kind='barh', color='skyblue')
plt.title('Top 20 Athletes with Multiple Medals')
plt.xlabel('Number of Medals')
plt.ylabel('Athlete')
plt.tight_layout()
plt.show()
```

Approximately 24.87 percent of athletes have won more than one medal at the Olympics. This percentage is almost at 25%, showing that almost a quarter of the individuals who have competed in the Olympics have won more than one medal, thus implying athletic accomplishment in a specific sport throughout more than one Olympic season. The visualization showcases the top 20 athletes and their number of medals, and these individuals have all won more than one medal, making them part of the 25 percent who have won multiple medals. Michael Phelps has won the most Olympic medals at over 25 medals won.
